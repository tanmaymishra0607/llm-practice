```python
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI
load_dotenv()
```

```python
llm = ChatGoogleGenerativeAI(model="gemma-4-31b-it")
response = llm.invoke("who said I love you first in world?")
print(response.text)
```

```python
prompt ="write a single tagline for a quirky coffee shop.Only output the tagline,nothing else"
llm = ChatGoogleGenerativeAI(model="gemma-4-31b-it",temperature=0)
response = llm.invoke(prompt)
print(response.text)
```
